# Basics of Python

## Expressions

An expression is a combination of values, variables, operators, and function calls that is evaluated to produce a result. Examples:

```python
print(15 ** 2)   # Use a '#' sign to write comments
```


## Variables

Variables are basically associations between a name and a value. The value is being stored in a location in the computer's memory. Examples:

```python
name = "Darwin"  #Â a string
age = 215        # a number
```

The two fundamental data types that the computer processes are:

1. Numbers (integers and floating-point numbers)
2. Strings (a sequence of characters)

All other data structures are built on top of these two fundamental data types, and are called *compound data types*.

Now that we have initialised our variables, we're free to use them in a computation by referring to them using their names:

```python
print(name)
print(age)
print(f"{name} is {age} years old")
```

## Control Flow

*Control flow* is a technical term we're using in programming to indicate a number of language constructs that guide the computer while performing a computation.

They fall within two great categories:

* Conditional statements
* Iteration statements.

A conditional statement allows the computer to do or *not do* something based on a condition we select. For example:

```python
if age > 18:
  print("The person is an adult")
```

For iteration (commonly referred to as *loops*) we're instructing a computer to perform an action repeatedly as long as a condition does or doesn't hold:

```python
baby_age = 1
while baby_age < 18:
  print(f"The baby is getting older: {baby_age}")
  baby_age += 1     # baby_age = baby_age + 1
```

## Collections

Collections allow us to bundle together related data and treat it as one unit.

The two most important collections in python are the `list` and the `dict`ionary.

A list allows us keep an arbitrary number of heterogeneous items together, for
example:

```python
my_favourite_things = ["Computers", "Piano", "Biology", 128]  # 4 items

# print collection
print(my_favourite_things)

print()

# print individual items in collection:
for item in my_favourite_things:
  print(item)
```

A list stores items at specific positions **(start from index 0)**.

```python
print(f"1st item: {my_favourite_things[0]}")

print()

# print individual items at each index in collection:
for idx in range(len(my_favourite_things)): # len: 4 (range : 0, 1, 2, 3)
  print(my_favourite_things[idx])
```

A dictionary allows us to keep arbitrary items but in a way that allows us to establish some associations between those (this is called a `key-value pair`):

```python
class_details = {
    "name": "How to Grow (Almost) Anything",
    "instructor": "David Kong",
    "institution": "MIT",
    "year": 2025
}

print(f"I <3 {class_details['name']} by {class_details['institution']}")
```

## Functions

Believe it or not, the above is more or less all you need to understand what a computer does from an abstract but fundamental level. Everything else builds on top of these ideas.

As an example, a *function* allows us to group a series of operations, give them a name, and treat the result as a functional unit, allowing us to incorporate this unit into our programs in a way that makes it easy to reuse the code.

To define a function in python, we use the `def` keyword followed by the name
of the function, and a pair open and closing parentheses:


```python
def bark():
  """
  Print a message. A dog barks.
  Return: None
  """
  print("woof!")

bark()

print()

bark()
```

Now, if you're wondering what's the point of the parentheses, they're there to allow us to pass in *arguments*, that is data for the function to use while it's performing the computation it performs:

```python
def dog_bark(color):
  """
  Return a message about the dog's color.
  """
  return f"{color} dog woof!"

brown_dog = dog_bark(color="brown")
print(brown_dog)

yellow_dog = dog_bark("yellow")
print(yellow_dog)
```

## Classes

Building up our programming toolkit, the next idea that helps us organise programs is the idea of a *Class*.

A class allows us to bundle together functions (called *methods* in this context) along with the data they operate on.

For example, for a rudimentary example of a class that represents a human:

```python
class Dog():
  def __init__(self, color, name):
    self.color = color
    self.name  = name
    # print("we initialize the special method __init__ here.")

  def dog_bark(self):
    print(f"I am a {self.color} dog named {self.name}. Woof!")

chill_dog = Dog(color="brown", name="chill_guy")
chill_dog.dog_bark()

### Reusability ###

nochill_dog = Dog("yellow", "no_chill_guy")
nochill_dog.dog_bark()
```

## Error bars using `matplotlib`, `numpy`

#### Visualize Uncertainty


Adding error bars lets a plot communicate not just "how big" an effect looks, but also "how reliable" that estimate is.



```python
# 8-well example: same mean, very different variability
control = np.array([9.9, 10.0, 10.1, 9.8, 10.0, 10.2, 9.9, 10.1])          # consistent wells
treatment = np.array([2.0, 6.0, 10.0, 14.0, 18.0, 10.0, 4.0, 16.0])        # noisy wells

groups = ["Control", "Treatment"]
means = [control.mean(), treatment.mean()]
stds = [control.std(ddof=1), treatment.std(ddof=1)]  # sample SD

plt.figure(figsize=(6, 4))
plt.bar(groups, means, yerr=stds, capsize=8)
plt.xlabel("Condition (8 wells each)")
plt.ylabel("Signal (a.u.)")
plt.title("Same mean, different variability")
plt.ylim(0, 22)
plt.show()
```

## Count Nucleotides in Strand

For this application we're going to look at counting the number of nucleotides (the basic blocks) of a DNA strand.

This is going to be our function:

```python
def count_nucleotides(dna_strand):
  strand_counts = {
      "a": 0,
      "t": 0,
      "c": 0,
      "g": 0
  }

  for nucleotide in dna_strand:
    strand_counts[nucleotide] += 1

  print("Number of nucleotides in strand: ")
  print(f"A: {strand_counts['a']}")
  print(f"T: {strand_counts['t']}")
  print(f"C: {strand_counts['c']}")
  print(f"G: {strand_counts['g']}")
```

Let's now use this function to count the number of nucleotides in a number of strands:

```python
first_strand = 'atgaatgcaaatga'
count_nucleotides(first_strand)

print()

second_strand = 'AGCTTTTCATTCTGACTGCAACGGGCAATATGTCTCTGTGTGGATTAAAAAAAGAGTGTCTGATAGCAGC'.lower()
count_nucleotides(second_strand)
```

## Find Common Kmers

Let's start with some applications to see how we can write a small program within the context of Bioinformatics to do something for us.

For our hello-world style, program, let's write a function that finds common kmers (polymers with variable number of monomers) inside a DNA molecule.

```python
def find_common_kmers(dna, k, threshold):
  ''' Find common kmers in argument @dna. Only reports
      kmers founds @threshold times.
  '''
  result = []
  strand_sequence = range(len(dna) - 1)

  for start in strand_sequence:
      kmer = dna[start : start + k]
      # print(f"kmer - dna[{start} : {start + k}]: {kmer}")
      if dna.count(kmer) >= threshold:
          result.append(kmer)

  return result
```

Before we start using our function, let's see real quick how it works, because it's a big jump from our basic definitions above.

Our function is called `find_common_kmers`, and takes 3 arguments (incoming data for processing):

1. a string representing a DNA strand, in the first argument `dna`,
2. the `k`-number we want that tunes the polymer length we're searching for,
3. an inclusion `threshold` - we're going to report the kmer as found if it
   exists at least `threshold` times.

Let's now start using our program. Let's ask it how many various polymers are found in the strand. Assume our strand is `atgaatgcaaatga`.

```python
dna_strand = 'atgaatgcaaatga'

# Show me the 2-mers that are present in the strand at least 2 times.
print("Common 2-mers:")
print(find_common_kmers(dna_strand, 2, 2))

print()

# Show me the 3-mers that are present in the strand at least 3 times.
print("Common 3-mers:")
print(find_common_kmers(dna_strand, 3, 3))
```
